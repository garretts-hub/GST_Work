# Problem: how to find the frequency of a discretely sampled biased sinusoid

Consider a qubit which is measured and found to be in the "1" state with probability

$p_1 = \frac{1}{2} + a \sin^2(\omega t)$

How can we best find $a$ and $\omega$?

```
import numpy as np
import matplotlib.pyplot as plt
import tensorflow  as  tf
import seaborn as sns
import pandas as pd
from scipy.optimize import curve_fit
```

## 1. Create discrete sampled biased sinusoid

```
def sample_sine(omega, nsamples_per_period=10, nperiods=100, amplitude=0.1):
  '''
  Compute nsamples_per_period * nperiods samples of 
  
      1/2 + a * sin(omega*t)
  
  discretized.
  
  The discretization is accomplished by letting the above expression be the probability p for 1.
  
  Return dict, with {"prob": probabilities, "samples": samples, 'tpts': time points}
  where values are numpy arrays.
  
  Include original values of omega and amplitude, for ground truth comparisons later.
  '''
  npts = nsamples_per_period * nperiods
  tpts = np.linspace(0, 2*np.pi/omega*nperiods, npts)
  sample_rate = 1/(tpts[2] - tpts[1])
  prob = (0.5 + amplitude * np.sin(tpts * omega))
  rand = np.random.random(npts)
  samples = (rand > 1-prob) * 1
  return {'prob': prob, "samples": samples, 'tpts': tpts, "omega": omega, "amplitude": amplitude, "sample rate": round(sample_rate,3)}
```

```
# plot one example dataset

omega = 1
samples_per_period = 5000
periods = 1
amplitude = 0.25
data = sample_sine(omega, samples_per_period, periods, amplitude)
samples = data['samples']
tpts = data['tpts']
plt.scatter(tpts, samples)
plt.show()
```

```
# recover sinusoid as a test, using a moving average
navg = 500
def moving_average(a, n=3) :
    ret = np.cumsum(a, dtype=float)
    ret[n:] = ret[n:] - ret[:-n]
    return ret[n - 1:] / n
  
plt.scatter(tpts[navg-1:], moving_average(samples, navg))
plt.show()
```

## Model and likelihood functions

```
# recover sinsuoid using fitting
def h1(x, theta):
    '''
    Generalized linear model function parametrized by theta.
    Returns probability of Y random variable being one, given the input value x.
    '''
    return 1/(1 + np.exp(-theta*x))
  
def h0(x, theta):
  '''
  Generalized linear model function parametrized by theta.
  Returns probability of random variable Y being in zero, givven input x.
  '''
  return 1-h1(x, theta)


def likelihood(x_array, y_array, theta):
  '''
  A likelihood function modeling the probability that Y is 0 or 1 given input X
  '''
  product = 1
  for i in range(len(x_array)):
    product *= h1(x_array[i], theta)**y_array[i]*h0(x_array[i], theta)**(1-y_array[i])
  return product

def log_likelihood(x_array, y_array, theta):
  '''
  The natural log of the likelihood function.
  '''
  summed = 0
  for i in range(len(x_array)):
    summed += np.ln(h1(x_array[i], theta)**y_array[i]*h0(x_array[i], theta)**(1-y_array[i]))
    
```

## Try using tensorflow for the ML regression


```
def model(data, nepochs=2, verbose=True):
  '''
  Tensorflow model of discrete sampled biased qubit evolution.
  Runs maximum likelihood fitting, returns omega, amplitude results.
  
  data = dict with data, and optionally the true omega, amplitude (for comparison)
  nepochs = number of "epochs" of fitting, where each epoch goes through all the data
  '''
  samples = data['samples']
  tpts = data['tpts']
  batch_size = len(tpts)
  
  # graph input
  X = tf.placeholder(tf.float32, name="time_points")
  Y = tf.placeholder(tf.float32, name="samples")
  
  # model variables
  omega = tf.Variable(1.0, name="omega")
  amplitude = tf.Variable(0.1, name="amplitude")
  
  # model to predict sample probability
  Y_sp = 0.5 + amplitude * tf.sin(omega * X)
  
  # loss function (binary cross-entropy)
  loss = - Y * tf.log(Y_sp) - (1-Y) * tf.log(1-Y_sp)
  
  # optimizer
  optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001).minimize(loss)
  
  if verbose:
    print("Running fitting over %s datapoints with %s epochs" % (batch_size, nepochs))
  with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for k in range(nepochs):
      for s, t in zip(samples, tpts):
        sess.run(optimizer, feed_dict={X:t, Y:s})
        
    results = sess.run([omega, amplitude])
  
  m_omega, m_amplitude = results
  if verbose:
    print("Results from ML regression: omega=%s, amplitude=%s" % (m_omega, m_amplitude))
    if 'omega' in data:
      rmse = [ np.sqrt((m_omega - data['omega'])**2), np.sqrt((m_amplitude - data['amplitude'])**2) ]
      print("Actual omega=%s, amplitude=%s, rmse=%s" % (data['omega'], data['amplitude'], rmse))
        
  return results
```

### Evaluate model performance vs number of samples per period

```
def eval_performance(omega, amplitude, n_data=2000, npts=4, nepochs=1, do_plot=True, verbose=True):
    '''
    Fit model for a range of n_samples_per_period, to see how the RMSE varys.
    
    Return Pandas data frame with the raw data.
    '''
    nspp_set = np.linspace(10, 2000, npts).astype(np.int32)
    rmse_set = []
    for nspp in nspp_set:
        nperiods = int(npts / nspp)+1
        data = sample_sine(omega, nspp, nperiods, amplitude=0.4)
        print("========== nspp=%s, nperiods=%s" % (nspp, nperiods))
        m_omega, m_amplitude = model(data, nepochs, verbose=verbose)    # do model fitting
        rmse = [ np.sqrt((m_omega - data['omega'])**2), np.sqrt((m_amplitude - data['amplitude'])**2) ]
        rmse_set.append(rmse)
    rmse_set = np.array(rmse_set)
    print("Done: have %d evaluation points" % nspp_set.shape[0])
    
    alldat = np.concatenate([nspp_set.reshape([-1, 1]), rmse_set], axis=1)
    df = pd.DataFrame(alldat, columns=["n_samples_per_period", "rmse_omega", "rmse_amplitude"])
    
    if do_plot:
        ax = df.plot(x="n_samples_per_period", y="rmse_omega", legend=False, figsize=(8,6), style=".-")
        ax2 = ax.twinx()
        df.plot(x="n_samples_per_period", y="rmse_amplitude", legend=False, color="r", ax=ax2, style=".-")
        lines1, labels1 = ax.get_legend_handles_labels()
        lines2, labels2 = ax2.get_legend_handles_labels()
        ax.figure.legend([lines1[0], lines2[0]], [labels1[0], labels2[0]])
        plt.show()
    if 0:
        # alternate style of plotting
        sns.set()
        ax = sns.pointplot(x="n_samples_per_period", y="rmse_omega", data=df)
        ax2 = sns.pointplot(x="n_samples_per_period", y="rmse_amplitude", data=df)
        
    return df
```

Observations from this numerical experiment:

- There appears to be a threshold for n_samples_per_period, above which the error drops sharply; for this example, that threshold is around nspp=500.  
- It would be good to see how the RMSE values behave with increases in the number of epochs used in the model fitting

Other comments:
